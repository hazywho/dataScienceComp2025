```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the data
data = pd.read_excel("Student-Employability-Datasets.xlsx", sheet_name="Data")

# Filter the data into two categories
employable = data[data["CLASS"] == "Employable"]
less_employable = data[data["CLASS"] == "LessEmployable"]

# Columns to consider for the highest score
columns = ["GENERAL APPEARANCE", "MANNER OF SPEAKING", "PHYSICAL CONDITION", 
           "MENTAL ALERTNESS", "SELF-CONFIDENCE", "ABILITY TO PRESENT IDEAS", 
           "COMMUNICATION SKILLS", "Student Performance Rating"]

# Function to count the highest scores
def count_highest_scores(df, columns):
    highest_counts = {col: 0 for col in columns}
    for index, row in df.iterrows():
        highest_score = row[columns].max()
        highest_col = row[columns][row[columns] == highest_score].index[0]
        highest_counts[highest_col] += 1
    return highest_counts

# Count highest scores for each category
employable_counts = count_highest_scores(employable, columns)
less_employable_counts = count_highest_scores(less_employable, columns)

# Plotting
fig, ax = plt.subplots(figsize=(12, 12))

# Bar positions
bar_width = 0.35
index = range(len(columns))

# Plot bars for Employable
bars1 = ax.bar(index, employable_counts.values(), bar_width, label="Employable")

# Plot bars for LessEmployable
bars2 = ax.bar([i + bar_width for i in index], less_employable_counts.values(), bar_width, label="Less Employable")

# Labels and title
ax.set_xlabel("Categories")
ax.set_ylabel("Number of Students")
ax.set_title("Highest Scores by Category")
ax.set_xticks([i + bar_width / 2 for i in index])
ax.set_xticklabels(columns, rotation=25, ha="right")
ax.legend()

plt.savefig(fname="employabilityComparison.jpg",dpi=300)
plt.show()
```

The dataset was collected from different university agencies in the Philippines.
