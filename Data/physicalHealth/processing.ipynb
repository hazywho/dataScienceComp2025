```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset (replace 'your_dataset.csv' with the actual file path)
data = pd.read_csv(r"C:\Users\zanyi\Documents\GitHub\dataScienceComp2025\Data\physicalHealth\Impact_of_Mobile_Phone_on_Students_Health.csv")

columns = ["Bad Health","Good Health"]

# Define criteria for "Health Benefit"
health_benefit_criteria = (
    (data["Mobile phone use for education"].isin(["Sometimes","Frequently"])) &  # Mobile phone use is helpful for studying
    (data["Health rating"].isin(["Poor"]))  # Health rating is Excellent or Good
)
# Categorize students
data["Health Benefit"] = health_benefit_criteria.map({True: "Good Health", False: "Bad Health"})
# Count the number of students in each category
health_benefit_counts = data["Health Benefit"].value_counts()
# Plot the bar chart
fig,ax = plt.subplots(figsize=(7,7))

# Bar positions
bar_width = 0.35
x = np.arange(len(columns)) *1.5



bars1 = ax.bar(x - bar_width/2, health_benefit_counts.values, width=bar_width, linewidth=2, zorder=5, label="Students using phone")
# Add annotations
# Add annotations
for bar in bars1:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, str(height), 
             ha='center', va='bottom', color='black', fontsize=12)
    
    
#normal
health_benefit_criteria = (
    (data["Mobile phone use for education"].isin(["Never","Rarely"])) &
    (data["Health rating"].isin(["Excellent","Good","Fair"]))  # Health rating is Excellent or Good
)
# Categorize students
data["Health Benefit"] = health_benefit_criteria.map({True: "Good Health", False: "Bad Health"})
# Count the number of students in each category
health_benefit_counts = data["Health Benefit"].value_counts()
bars2 = ax.bar(x + bar_width/2, health_benefit_counts.values, width=bar_width, linewidth=2, zorder=5, label="Normal students")
# Add annotations
for bar in bars2:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, str(height), 
             ha='center', va='bottom', color='black', fontsize=12)
    
    
# Add labels and title
plt.xticks(x, columns, fontsize=12, fontweight='bold', color='black')  # Fix: Set tick labels 
plt.xlabel("Diagnosis")
plt.ylabel("Number of Students")
plt.title("Comparison of using devices for studies")
ax.legend()

plt.savefig(fname="testing.jpg",dpi=300)
# Show the plot
plt.show()
```
