```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"SATSCORE.csv")
df.head()
```

```python
print(df.shape)
df.isna().sum()
```

```python
df_sorted=df.sort_values("Year")
df_sorted
```

```python
fig,ax = plt.subplots(1,1)

ax.plot(df_sorted["Year"],df_sorted["M_averages"],linestyle="-",color="blue",label="male")
ax.plot(df_sorted["Year"],df_sorted["F_averages"],linestyle="-",color="orange",label="female")
plt.ylabel("Average score")
plt.xlabel("Year")
plt.title("Average SAT score over the years")
plt.legend()

plt.savefig(fname="satScoreOverYears.jpg",dpi=300)
```
